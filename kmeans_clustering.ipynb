```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Lists to store SSE (inertia) and silhouette scores
cluster_inertia = []
s_score = []

for i in range(2, 14):
    kmeans = KMeans(n_clusters=i, random_state=42, max_iter=1500)
    cluster_labels = kmeans.fit_predict(df.iloc[:, 2:5])  # Get cluster labels
    cluster_inertia.append(kmeans.inertia_)
    s_score.append(silhouette_score(df.iloc[:, 2:5], cluster_labels))  # Calculate silhouette score

# Plotting the results
plt.figure(figsize=(12, 5))

# Elbow method plot for SSE (Inertia)
plt.subplot(1, 2, 1)
plt.plot(range(2, 14), cluster_inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Cluster Inertia')

# Silhouette score plot
plt.subplot(1, 2, 2)
plt.plot(range(2, 14), s_score, color='orange', marker='o')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')

plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=7, random_state=42, max_iter=1500)
cluster_labels = kmeans.fit_predict(df.iloc[:, 2:5])
df['Cluster'] = cluster_labels
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Annual Income (k$)'], y=df['Spending Score (1-100)'], hue=df['Cluster'], palette='viridis')
plt.title('Customer Segments Based on Annual Income and Spending Score')
plt.show()
```
