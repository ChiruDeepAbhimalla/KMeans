```python
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
d_img=mpl.image.imread('/content/dog.jpg')
```

```python
plt.imshow(d_img)
```

```python
d_img.shape
```

```python
x=d_img.reshape(-1,3)
x.shape
```

```python
kmeans=KMeans(n_clusters=20)
```

```python
kmeans.fit(x)
```

```python
segmented_image=kmeans.cluster_centers_[kmeans.labels_]
segmented_image=segmented_image.reshape(d_img.shape)
```

```python
plt.imshow(segmented_image/255)
```

```python
import cv2

cv2.imwrite('image1.png',cv2.cvtColor(d_img,cv2.COLOR_BGR2RGB))

cv2.imwrite('image2.png',cv2.cvtColor(segmented_image.astype("uint8"),cv2.COLOR_BGR2RGB))
```

```python
fl_img=mpl.image.imread('/content/flower.jpg')
```

```python
plt.imshow(fl_img)
```

```python
fl_img.shape
```

```python
y=fl_img.reshape(-1,3)
y.shape
```

```python
kmeans_f=KMeans(n_clusters=10)
```
